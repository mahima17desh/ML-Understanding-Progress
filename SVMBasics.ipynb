```python
"""
We will predict whether cancer is Benign or Malignant using historical data
about patients diagnosed with cancer.
"""
#Loads the breast cancer dataset
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
#To automatically draw the decision boundary of the svm model
from sklearn.inspection import DecisionBoundaryDisplay
#Allows to create and train an SVM classifier.
from sklearn.svm import SVC

cancer = load_breast_cancer()
X = cancer.data[:, :2]
#Two features are chosen because we can only visualize data in 2D.
#Target labels:0 = Malignant, 1 = Benign
y = cancer.target

#Creates a Support Vector Classifier with a linear kernel and regularization parameter C=1.
svm = SVC(kernel="linear", C=1)
#Trains the SVM model on the feature matrix X and target labels y.
svm.fit(X, y)

# Visualizes the decision boundary of the trained model with a specified color map.
DecisionBoundaryDisplay.from_estimator(
        svm,
        X,
        response_method="predict",
        alpha=0.8,
        cmap="Pastel1",
        xlabel=cancer.feature_names[0],
        ylabel=cancer.feature_names[1],
    )

#Plotting
plt.scatter(X[:, 0], X[:, 1],
            c=y,
            s=20, edgecolors="k")
#Displaying the support vectors
plt.scatter(
    svm.support_vectors_[:, 0],
    svm.support_vectors_[:, 1],
    s=120,
    facecolors="none",
    edgecolors="red",
    linewidth=2,
    label="Support Vectors"
)

plt.legend()
plt.show()
```
